```python
import cv2
import numpy as np

def undistort_fisheye(input_image_path):
    # 카메라 파라미터 설정 (K: 카메라 매트릭스, D: 왜곡 계수)
    K = np.array([[640, 0, 320], [0, 640, 240], [0, 0, 1]], dtype=np.float32)
    D = np.array([0, 0, 0, 0], dtype=np.float32)

    # 입력 이미지 읽기
    img = cv2.imread(input_image_path)

    # 왜곡 보정
    undistorted_img = cv2.fisheye.undistortImage(img, K, D)
    return undistorted_img

undistorted_img = undistort_fisheye(input_image_path)




fig = plt.figure(figsize=(16,8))
plt.imshow(undistorted_img)
plt.show()
```

```python
def apply_inverse_fisheye_distortion(image_path, mask=False):

    # 이미지 불러오기
    if mask == True:
        image = cv2.imread(image_path, 0)
    else:
        image = cv2.imread(image_path)

    # 이미지 크기 가져오기
    height, width = image.shape[:2]

    # 카메라 매트릭스 생성 (원본 왜곡을 생성했던 것과 동일)
    focal_length = width / 4
    center_x = width / 2
    center_y = height / 2
    camera_matrix = np.array([[focal_length, 0, center_x],
                              [0, focal_length, center_y],
                              [0, 0, 1]], dtype=np.float32)

    # 왜곡 계수 생성 (원본 왜곡 계수와 동일)
    dist_coeffs = np.array([10, 1, 0, 0], dtype=np.float32)

    # 역 왜곡 보정
    undistorted_image = cv2.undistort(image, camera_matrix, dist_coeffs, None, camera_matrix)

    return undistorted_image

# 역 왜곡 적용
undistorted_img = apply_inverse_fisheye_distortion(input_image_path)
undistorted_mask = apply_inverse_fisheye_distortion(mask_image_path, mask=True)

fig = plt.figure(figsize=(16,8))
plt.imshow(undistorted_img)
plt.show()
```

```python
def apply_fisheye_distortion(image_path, mask=False):

    # 이미지 불러오기
    if mask==True:
        image = cv2.imread(image_path, 0)
    else:
        image = cv2.imread(image_path)

    # 이미지 크기 가져오기
    height, width = image.shape[:2]

    # 카메라 매트릭스 생성
    focal_length = width / 4
    center_x = width / 2
    center_y = height / 2
    camera_matrix = np.array([[focal_length, 0, center_x],
                              [0, focal_length, center_y],
                              [0, 0, 1]], dtype=np.float32)

    # 왜곡 계수 생성
    dist_coeffs = np.array([0, -0.1, 0, 0], dtype=np.float32)

    # 왜곡 보정
    undistorted_image = cv2.undistort(image, camera_matrix, dist_coeffs)

    return(undistorted_image)
    # 피셔 아이 왜곡 적용
fisheye_img = apply_fisheye_distortion(input_image_path)
fisheye_mask = apply_fisheye_distortion(mask_image_path, mask=True)

fig = plt.figure(figsize=(16,8))
plt.imshow(fisheye_img)
plt.show()
```
